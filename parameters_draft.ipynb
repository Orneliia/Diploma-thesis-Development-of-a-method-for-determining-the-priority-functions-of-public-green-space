```python
# Получить список всех файлов в папке (например, GeoJSON)
service_files = [os.path.join(services_folder, f) for f in os.listdir(services_folder) if f.endswith(".geojson")]
```

```python
# Убедиться, что CRS для парков установлен
if parks_gdf.crs.to_string() != "EPSG:32636":
    parks_gdf = parks_gdf.to_crs("EPSG:32636")
```

```python
# Инициализация результирующего столбца для подсчёта услуг
parks_gdf["services_count"] = 0
```

```python
def process_service_file(service_file, parks_gdf):
    """
    Обрабатывает один файл с сервисами, добавляет их в зоны парков.
    """
    # Загрузить данные о сервисах
    services_gdf = gpd.read_file(service_file)
    
    # Убедиться, что CRS совпадает с парками
    if services_gdf.crs.to_string() != "EPSG:32636":
        services_gdf = services_gdf.to_crs("EPSG:32636")
    
    # Привязать точки сервисов к полигонам парков
    joined = gpd.sjoin(
        services_gdf, parks_gdf, how="inner", predicate="within", 
        lsuffix="service", rsuffix="park"
    )
    
    # Подсчитать количество сервисов для каждого парка
    service_counts = joined.groupby("index_park").size()
    
    # Добавить эти значения в DataFrame парков
    for index, count in service_counts.items():
        parks_gdf.loc[index, "services_count"] += count
    
    return parks_gdf

```

```python
# Обработка каждого файла
for service_file in service_files:
    parks_gdf = process_service_file(service_file, parks_gdf)
```
